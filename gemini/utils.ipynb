# Reformat

```python
import json

with open('test_data.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

converted = []
for item in data:
    converted.append({
        "id": item["id"],
        "nat_premises": item["premises"],
        "nat_conclusion": item["conclusion"],
        "label": item["label"]
    })

with open('test_data.json', 'w', encoding='utf-8') as f:
    json.dump(converted, f, ensure_ascii=False, indent=4)
```

# Compare predictions & true labels

```python
'''Kiếm tra có bao nhiêu predictions khớp và không khớp true labels'''
import json

prediction_file = "predictions.json" # <-- Sửa ở đây (nếu cần), đây là file chứa predictions 
data_file = "test_data.json" # <-- Sửa ở đây (nếu cần), đây là file chứa dataset test gốc

with open(prediction_file, 'r', encoding='utf-8') as f:
    predictions = json.load(f)

with open(data_file, 'r', encoding='utf-8') as f:
    test_data = json.load(f)

true_labels = {s['id'] : s['label'] for s in test_data}
# true_labels = {s['id'] : s['label'] for s in test_data if int(s['id'][7:]) in range(00, 500)}

correct = 0
incorrect = 0
total = len(true_labels)
for k in true_labels:
    if predictions[k] == true_labels[k]:
        correct += 1
    else:
        incorrect += 1

print("Total:", total)
if (total):
    print(f"- Correct predictions: {correct}/{total} ({correct/total * 100} %)")
    print(f"- Incorrect predictions: {incorrect}/{total} ({incorrect/total * 100} %)")
```

```python
visited_ids = []
label_ids = [d['id'] for d in test_data]

count = 0
for id in label_ids:
    if (id not in visited_ids):
        visited_ids.append(id)
    else:
        print(id)

print(count)
```
