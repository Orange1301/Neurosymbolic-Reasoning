# Re-format dataset

```python
import json
```

```python
def write_file(data, file_path: str):
    with open(file_path, "w", encoding="utf-8") as f:
        json.dump(data, f, ensure_ascii=False, indent=4)

def reformat_dataset(file_path: str, id_str: str):
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    sorted_data = sorted(data, key=lambda x: x["story_id"])

    cur_index = 0
    formatted_data = []
    for d in data:
        nat_premises, nat_conclusion = d["natural"].rsplit("\n", 1)
        fol_premises, fol_conclusion = d["fol"].rsplit("\n", 1)

        formatted_data.append({
            "id": id_str + '_' + str(cur_index),
            "story_id": d["story_id"],
            "nat_premises": nat_premises,
            "nat_conclusion": nat_conclusion,
            "fol_premises": fol_premises,
            "fol_conclusion": fol_conclusion,
            "label": d["label"]
        })

        cur_index += 1
    
    return formatted_data
```

```python
formatted_data = reformat_dataset("folio_train.json", "folio_train")
write_file(formatted_data, "new_folio_train.json")

formatted_data = reformat_dataset("folio_valid.json", "folio_valid")
write_file(formatted_data, "new_folio_valid.json")

formatted_data = reformat_dataset("folio_test.json", "folio_test")
write_file(formatted_data, "new_folio_test.json")
```
