```python
from datasets import load_dataset
import json
```

```python
dataset = load_dataset("datatune/LogiQA2.0")

for split in dataset.keys():
    save_path = f"raw/raw_logiqa_{split}.json"
    data_list = dataset[split].to_list()

    raw_data = []
    for d in data_list:
        try:
            json_string = d["text"].strip()
            parsed_item = json.loads(json_string)
            
            raw_data.append(parsed_item)
        except json.JSONDecodeError as e:
            continue

    with open(save_path, "w", encoding="utf-8") as f:
        json.dump(raw_data, f, ensure_ascii=False, indent=4)
```
