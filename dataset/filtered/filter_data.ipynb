# Filter data

```python
import json
import numpy as np
import random

dst_file = 'reclor_valid.json'
with open('../full/ReClor/reclor_valid.json') as f:
    data = json.load(f)

data = sorted(data, key=lambda x: len(x['nat_premises']))

# Split T-F-U
true_samples = []
false_samples = []
uncertain_samples = []

for d in data:
    if d["label"] == "True":
        true_samples.append(d)
    elif d["label"] == "False":
        false_samples.append(d)
    else:
        uncertain_samples.append(d)

# Filter 
new_N = 80

N = len(true_samples)
indices = np.linspace(0, N - 1, new_N, dtype=int)
true_filtered = [true_samples[i] for i in indices]

N = len(false_samples)
indices = np.linspace(0, N - 1, new_N, dtype=int)
# false_filtered = [false_samples[i] for i in indices]
false_filtered = []

# Save
filtered_samples = true_filtered + false_filtered + uncertain_samples
random.shuffle(filtered_samples)
with open(dst_file, 'w', encoding='utf-8') as f:
    json.dump(filtered_samples, f, ensure_ascii=False, indent=4)
```
