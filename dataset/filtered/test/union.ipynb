# Đánh id lại theo thứ tự

```python
import json

manual_files = ['./Bao_manual.json', 'Khoi_manual.json']

data = []
for file in manual_files:
    with open(file, 'r', encoding='utf-8') as f:
        data.extend(json.load(f))

for i, d in enumerate(data):
    d['id'] = f'manual_{i}'

with open('final_manual_test.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=4)
```

# Đếm T, F, U

```python
import json

with open('Bao_manual.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

t = f = u = 0
for d in data:
    if d["label"] == "True": t += 1
    elif d["label"] == "False": f += 1
    else: u += 1

print(t, f, u)
```
