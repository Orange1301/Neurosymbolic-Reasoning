```python
import json

manual_test_name = "test/manual_test.json"
folio_valid_name = "valid/folio_valid.json"

with open(manual_test_name, "r", encoding="utf-8") as f:
    manual_data = json.load(f)
    
with open(folio_valid_name, "r", encoding="utf-8") as f:
    folio_data = json.load(f)
```

```python
valid_data = []
for data in folio_data:
    valid_data.append({
        "id": data["id"],
        "premises": data["nat_premises"],
        "conclusion": data["nat_conclusion"],
        "label": data["label"]
	})
```

```python
import random

random.shuffle(manual_data)
valid_data.extend(manual_data[:100])
manual_data = manual_data[100:]
```

```python
valid_name = "valid/valid.json"
manual_name = "test/manual_preprocess_test.json"

with open(valid_name, "w", encoding="utf-8") as f:
    json.dump(valid_data, f, indent=4, ensure_ascii=False)
with open(manual_name, "w", encoding="utf-8") as f:
    json.dump(manual_data, f, indent=4, ensure_ascii=False)
```
