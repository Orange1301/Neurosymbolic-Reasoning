```python
import json

# folio_train_name = "train/folio_train.json"
logiqa_train_name = "train/logiqa_train.json"
proofwriter_train_name = "train/proofwriter_train.json"
reclor_train_name = "train/reclor_train.json"
ruletaker_train_name = "train/ruletaker_train.json"

train_name_arr = [
    logiqa_train_name, 
    proofwriter_train_name, 
    reclor_train_name, 
    ruletaker_train_name
]
```

```python
all_data = []

for train_name in train_name_arr:
	with open(train_name, 'r', encoding="utf-8") as f:
		data_one_file = json.load(f)
		all_data.extend(data_one_file)
```

```python
for data in all_data:
    premises = data["nat_premises"].replace('.', '').split()
    conclusion = data["nat_conclusion"].replace('.', '').split()
    data["length"] = len(premises) + len(conclusion)
```

```python
all_data.sort(key=lambda x: x["length"], reverse=False)
```

```python
merge_train_name = "train/train.json"
with open(merge_train_name, 'w', encoding="utf-8") as f:
    json.dump(all_data, f, indent=4, ensure_ascii=False)
```
