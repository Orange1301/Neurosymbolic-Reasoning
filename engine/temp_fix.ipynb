# Re-format wrong files

```python
import json

def write_file(data, file_path: str):
    with open(file_path, "w", encoding="utf-8") as f:
        json.dump(data, f, ensure_ascii=False, indent=4)

def reformat_wrong_files(file_path: str):
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    formatted_data = []
    for d in data:
        natural = (d["nat_premises"] + "\n" + d["nat_conclusion"]).split('\n')
        fol = (d["fol_premises"] + "\n" + d["fol_conclusion"]).split('\n')
        formatted_data.append({
            "id": d["id"],
            "natural": natural,
            "fol": fol,
            "label": d["label"]
        })
    
    write_file(formatted_data, file_path)
```

```python
reformat_wrong_files("wrong_folio_train.json")
reformat_wrong_files("wrong_folio_valid.json")
reformat_wrong_files("wrong_folio_test.json")
```
